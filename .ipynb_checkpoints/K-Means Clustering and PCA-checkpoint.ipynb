## Preprocessing

```python
# Get dummies for categorical features
data = pd.get_dummies(df_dropped)
print(data.shape)
```

```python
X = data.drop(['hotel_cluster'], axis=1)
Y = data['hotel_cluster']
```

```python
# Select top K features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

n = 8
selector = SelectKBest(chi2, k=n)
X_new = selector.fit_transform(X, Y)

print(X_new.shape)
#print(X_new.columns)

mask = selector.get_support()
new_features = X.columns[mask]
print(new_features)
```

```python
from sklearn.model_selection import train_test_split

# Normal data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

# Feature selection data
#X_train, X_test, y_train, y_test = train_test_split(X_new, Y, test_size=0.25, random_state=42)

print(X_train.shape)
```

## K-Means Clustering

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score, roc_curve, log_loss, average_precision_score

def multiclass_roc_auc_score(y_test, y_pred):
    lb = LabelBinarizer()
    lb.fit(y_test)
    y_test = lb.transform(y_test)
    y_pred = lb.transform(y_pred)
    return roc_auc_score(y_test, y_pred)

def evaluate(y_test, y_test_pred, y_test_pred_proba):
    accuracy = accuracy_score(y_test, y_test_pred)
    print('Accuracy', accuracy)

    #auc = roc_auc_score(y_test, y_test_pred, multi_class='ovr')
    auc = multiclass_roc_auc_score(y_test, y_test_pred)
    
    #binary auc
    #auc = roc_auc_score(y_test, y_test_pred_proba[:,1])

    print('AUC', auc)

    cross_entropy = log_loss(y_test, y_test_pred_proba)
    print('Cross entropy', cross_entropy)
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans

#Use Elbow method to determine number of clusters
wcss = []
for i in range(1, 25):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 25), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=100, max_iter = 10000, tol = 0.0000001).fit(X_train)

y_test_pred = kmeans.predict(X_test)

# Centroid values
centroids = kmeans.cluster_centers_

accuracy = accuracy_score(y_test, y_test_pred)
print('Accuracy', accuracy)

auc = multiclass_roc_auc_score(y_test, y_test_pred)
print('AUC', auc)

print(centroids)
```

## Catboost

```python
# Catboost
from catboost import Pool, CatBoostClassifier

# Indices of categorical features
cat_features = [0,1,2,3,5,6,7,11,12,13,15,16,17]


train_dataset = Pool(data=X_train,
                     label=y_train,
                     cat_features=cat_features)

test_dataset = Pool(data=X_test,
                    label=y_test,
                    cat_features=cat_features)


# Initialize CatBoostClassifier
model = CatBoostClassifier(classes_count=100, loss_function="MultiClass", n_estimators=10, verbose=2)

# Fit model
model.fit(train_dataset)
y_test_pred = model.predict(test_dataset)
y_test_pred_proba = model.predict_proba(test_dataset)

evaluate(y_test, y_test_pred, y_test_pred_proba)
```
